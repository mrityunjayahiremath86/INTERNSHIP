```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the ICC ODI team rankings page
url = 'https://www.icc-cricket.com/rankings/mens/team-rankings/odi'

# Send a GET request to the URL and store the response in a variable
response = requests.get(url)

# Parse the HTML content of the response using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table containing the team rankings and store it in a variable
table = soup.find('table', {'class': 'table'})

# Find all the rows in the table and store them in a list
rows = table.find_all('tr')

# Create an empty list to store the extracted data
data = []

# Iterate over the rows and extract the data
for row in rows[1:21]:  # Start at index 1 to skip the table header row, and stop at index 21 to get the top 20 teams only
    cells = row.find_all('td')
    rank = cells[0].text.strip()
    team_name = cells[1].text.strip()
    matches = cells[2].text.strip()
    points = cells[3].text.strip()
    rating = cells[4].text.strip()
    data.append({'Rank': rank, 'Team': team_name, 'Matches': matches, 'Points': points, 'Rating': rating})

# Create a pandas dataframe from the extracted data
df = pd.DataFrame(data)

# Print the dataframe
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the CNBC World News page
url = 'https://www.cnbc.com/world/?region=world'

# Send a GET request to the URL and store the response in a variable
response = requests.get(url)

# Parse the HTML content of the response using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find all the articles on the page and store them in a list
articles = soup.find_all('div', class_="LatestNews-isHomePage LatestNews-isIntlHomepage")
articles

# Create an empty list to store the extracted data
data2 = []

# Iterate over the articles and extract the headline, time, and news link
for article in articles:
    headline = article.find('a').text.strip()
    time = article.find('time').text.strip()
    news_link = article.find('a')['href']
    data2.append({'Headline': headline, 'Time': time, 'News Link': news_link})

# Create a pandas dataframe from the extracted data
df2 = pd.DataFrame(data2)

# Print the dataframe
print(df2.head(5))
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime, timedelta

# Set the URL of the page to scrape
url = "https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"

# Define the number of days to look back for most downloaded articles
days_to_look_back = 90

# Get the date 90 days ago from today
start_date = datetime.today() - timedelta(days=days_to_look_back)

# Get the HTML content of the page
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Find the articles on the page
articles = soup.find_all("h4", class_="sc-9zxyh7-1 sc-9zxyh7-2 kOEIEO hvoVxs")
articles

# Create empty lists to store data
paper_titles = []
authors = []
published_dates = []
paper_urls = []

# Loop through the articles and extract the details
for article in articles:
    # Get the published date of the article
    published_date_str = article.find("div", class_="pod-listing-header").find("span", class_="pod-listing-header-pubdates").text.strip()
    published_date = datetime.strptime(published_date_str, "%d %B %Y").date()
    
    # Check if the article was published within the last 90 days
    if published_date >= start_date:
        # Get the paper title
        paper_title = article.find("h4", class_="pod-listing-title").text.strip()
        paper_titles.append(paper_title)
        
        # Get the authors
        author_tags = article.find_all("a", class_="pod-listing-author-link")
        author_names = [author_tag.text.strip() for author_tag in author_tags]
        authors.append(", ".join(author_names))
        
        # Get the published date
        published_dates.append(published_date_str)
        
        # Get the paper URL
        paper_url = "https://www.journals.elsevier.com" + article.find("a", class_="pod-listing-header-link")["href"]
        paper_urls.append(paper_url)

# Create a dataframe with the scraped data
data = {"Paper Title": paper_titles, "Authors": authors, "Published Date": published_dates, "Paper URL": paper_urls}
df = pd.DataFrame(data)

# Print the dataframe
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Send a GET request to the URL
url = 'https://www.dineout.co.in/delhi-restaurants/buffet-special'
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Extract restaurant details
restaurant_list = soup.find_all('div', class_='restnt-info-wrap')

# Initialize empty lists to store the scraped data
restaurant_name_list = []
cuisine_list = []
location_list = []
rating_list = []
image_url_list = []

# Loop through each restaurant and extract the required details
for restaurant in restaurant_list:
    # Extract restaurant name
    restaurant_name = restaurant.find('a', class_='restnt-name').text.strip()
    restaurant_name_list.append(restaurant_name)
    
    # Extract cuisine
    cuisine = restaurant.find('span', class_='double-line-ellipsis').text.strip()
    cuisine_list.append(cuisine)
    
    # Extract location
    location = restaurant.find('div', class_='restnt-loc ellipsis').text.strip()
    location_list.append(location)
    
    # Extract rating
    rating = restaurant.find('span', class_='rating-val').text.strip()
    rating_list.append(rating)
    
    # Extract image URL
    image_url = restaurant.find('img', class_='no-resize').get('data-original')
    image_url_list.append(image_url)

# Create a dictionary to store the scraped data
data_dict = {
    'Restaurant Name': restaurant_name_list,
    'Cuisine': cuisine_list,
    'Location': location_list,
    'Rating': rating_list,
    'Image URL': image_url_list
}

# Create a Pandas DataFrame using the data dictionary
df = pd.DataFrame(data_dict)

# Print the DataFrame
print(df)
```
